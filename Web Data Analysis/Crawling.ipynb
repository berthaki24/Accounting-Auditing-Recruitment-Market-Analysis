```python
import requests
import time

base_url = "https://internal-api.careerviet.vn/api/v1/js/jsk/jobs/public"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

keywords = [
    "accountant",
    "accounting",
    "general accountant",
    "senior accountant",
    "chief accountant",
    "audit",
    "auditor",
    "internal auditor",
    "audit assistant",
    "tax",
    "tax accountant",
    "finance",
    "financial analyst",
    "kế toán",
    "kế toán tổng hợp",
    "kế toán thuế",
    "kiểm toán"
]

all_jobs = []

for keyword in keywords:

    page = 1

    while True:

        params = {
            "page": page,
            "industry": 2,
            "keyword": keyword,
            "limit": 50
        }

        res = requests.get(base_url, headers=headers, params=params)

        if res.status_code != 200:
            print("Blocked")
            break

        data = res.json()
        jobs = data.get("data", [])

        if not jobs:
            break

        all_jobs.extend(jobs)

        print(f"{keyword} | page {page} | total {len(all_jobs)}")

        page += 1
        time.sleep(1)

print("Total jobs:", len(all_jobs))
```

```python
from bs4 import BeautifulSoup
import time
import requests
import re

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

def get_job_detail(job_id):
    url = f"https://internal-api.careerviet.vn/api/v1/js/jsk/jobs/{job_id}"
    res = requests.get(url, headers=headers)

    if res.status_code != 200:
        print("Error while fetching details:", job_id)
        return None

    return res.json().get("data", {})
```

```python
def get_job_detail(job_link):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Accept-Language": "vi-VN,vi;q=0.9"
    }

    job_link = job_link.replace("/en/", "/vi/")

    try:
        res = requests.get(job_link, headers=headers)

        print("Status:", res.status_code)

        if res.status_code != 200:
            print("Error while fetching details:", job_link)
            return "", ""

        soup = BeautifulSoup(res.text, "html.parser")

        sections = soup.find_all("div", class_="detail-row reset-bullet")

        description = ""
        requirement = ""

        if len(sections) >= 1:
            description = sections[0].get_text(strip=True)

        if len(sections) >= 2:
            requirement = sections[1].get_text(strip=True)

        return description, requirement

    except Exception as e:
        print("Exception:", e)
        return "", ""
```

```python
import requests
from bs4 import BeautifulSoup
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

session = requests.Session()
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}

def extract_simple_content(url):

    url = url.replace("/en/", "/vi/")

    result = {
        "job_description": " ",
        "requirements": " ",
        "address": " "
    }

    try:

        res = session.get(url, headers=headers, timeout=10)
        if res.status_code != 200:
            return result

        soup = BeautifulSoup(res.text, "html.parser")

        for r in soup(["script", "style", "nav", "footer"]):
            r.decompose()

        all_text = soup.get_text(separator="\n")
        lines = [line.strip() for line in all_text.split("\n") if line.strip()]

        current_section = None

        for line in lines:

            line_upper = line.upper().replace(":", "").strip()

            if line_upper == "MÔ TẢ CÔNG VIỆC":
                current_section = "job_description"
                result[current_section] = ""
                continue

            elif line_upper == "YÊU CẦU CÔNG VIỆC":
                current_section = "requirements"
                result[current_section] = ""
                continue

            elif line_upper == "ĐỊA ĐIỂM LÀM VIỆC":
                current_section = "address"
                result[current_section] = ""
                continue

            if line_upper in [
                "CÁCH THỨC ỨNG TUYỂN",
                "NGÀNH NGHỀ",
                "VIỆC LÀM CÙNG CÔNG TY",
                "VIỆC LÀM TƯƠNG TỰ"
            ]:
                current_section = None

            if current_section is not None:
                if "Click here to see" not in line:
                    result[current_section] += line + "\n"

        for key in result:
            if result[key] != " ":
                result[key] = result[key].strip()
                if not result[key]:
                    result[key] = " "

        return result

    except Exception as e:
        print(f"Error while crawling link {url}: {e}")
        return result


def process_job(job):

    job_link = job.get("job_link")

    detail_data = extract_simple_content(job_link)

    job_info = {
        "job_id": job.get("job_id"),
        "job_title": job.get("job_title"),
        "experience_min": job.get("job_experience"),
        "experience_max": job.get("job_to_experience"),
        "salary_min": job.get("job_from_salary"),
        "salary_max": job.get("job_to_salary"),
        "salary_string": job.get("job_salary_string"),
        "location": ", ".join(job.get("location_name_en", [])),
        "address": detail_data["address"],
        "job_active_date": job.get("job_active_date"),
        "link": job_link,
        "job_description": detail_data["job_description"],
        "requirements": detail_data["requirements"]
    }

    return job_info


test_jobs = all_jobs
extracted_data_final = []

print(f"Starting to crawl {len(test_jobs)} jobs")

workers = 15
pool_maxsize = 50

with ThreadPoolExecutor(max_workers=workers) as executor:

    futures = [executor.submit(process_job, job) for job in test_jobs]

    for i, future in enumerate(as_completed(futures)):
        result = future.result()
        extracted_data_final.append(result)

        if i % 50 == 0:
            print(f"Crawled {i} jobs")

print("Crawling completed")
```

```python
import csv

keys = extracted_data_final[0].keys()

with open("Tasks WDA - RAW DATA.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.DictWriter(f, fieldnames=keys)
    writer.writeheader()
    writer.writerows(extracted_data_final)

print("Saved successfully")
```
